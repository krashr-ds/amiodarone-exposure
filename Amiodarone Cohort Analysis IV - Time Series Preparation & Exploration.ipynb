## An OMOP Cohort: Patients Exposed to Amiodarone Hydrochloride > 30 Days

### Part IV: A Longitudinal Time Series Preparation & Exploration

An examination of de-identified longitudinal visit and procedure data (derived to OMOP from Medicare DeSYNPuf data, 2008-2010, by OHDSI) and associated ICD9CM diagnosis codes to analyze the incidence and prevalence of AEs associated with amiodarone toxicity to the lung, thyroid and liver.

```python
%matplotlib inline
import random
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
condition_visits_ds = pd.read_csv("conditions_visits_procedures.csv", parse_dates=["treatment_start", "vstart", 
                                                                                   "condition_start_date", "condition_end_date"])
```

Calculate treatment end-dates.

```python
condition_visits_ds['treatment_period'] = condition_visits_ds['treatment_period'].astype('float64')
temp = condition_visits_ds['treatment_period'].apply(np.ceil).apply(lambda x: pd.Timedelta(x, unit='D'))
condition_visits_ds['treatment_end'] = condition_visits_ds['treatment_start'] + temp
```

```python
condition_visits_ds.loc[condition_visits_ds["vstart"].between(condition_visits_ds["treatment_start"], condition_visits_ds["treatment_end"]), "in_period"] = 1
condition_visits_ds["in_period"] = condition_visits_ds["in_period"].fillna(0)
condition_visits_ds["in_period"] = condition_visits_ds["in_period"].astype("int64")

# sentinel_visits_ds.head()
```

```python
condition_visits_ds.loc[condition_visits_ds["vstart"] > condition_visits_ds["treatment_end"], "after_period"] = 1
condition_visits_ds["after_period"] = condition_visits_ds["after_period"].fillna(0)
condition_visits_ds["after_period"] = condition_visits_ds["after_period"].astype("int64")
```

Create ts dataframe, to look at average # of diagnoses, procedures, visits and sentinel_visits by year, month.

```python
condition_visits_ts = condition_visits_all_agg.groupby(["vyear", "vmonth"]).agg({'treatment_period': 'mean', 
                                                                            'visit_id': 'mean',
                                                                            'vprovider': 'mean',
                                                                            'care_site': 'mean',
                                                                            'sentinel_condition_code' : 'mean',
                                                                            'sentinel_condition_type' : 'mean',
                                                                            'visit_diagnoses' : 'mean',
                                                                            'visit_procedures' : 'mean'})

condition_visits_ts.columns = ["avg_treatment_period", "avg_visit_count", "avg_provider_count", "avg_site_count", 
                               "avg_sentinel_diagnoses", "avg_sentinel_types", "avg_visit_diagnoses", "avg_visit_procedures"]

condition_visits_ts = condition_visits_ts.round(2)
condition_visits_ts.head()
```
